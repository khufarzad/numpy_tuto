```python
import numpy as np
```

```python
def IQR_outlier_detection(v):
    q1, q3 = np.percentile(v, [25, 75])
    print(f'q1={q1}, q2={np.median(v)} ,q3={q3}')
    IQR = q3 - q1
    print('IQR=', IQR)
    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR
    print(f'lower={lower_bound}, upper={upper_bound}')
    return np.where((v > upper_bound) | (v < lower_bound))
```

```python
np.random.seed(100)
arr = np.append(np.random.rand(1000), np.array([20]))
out = IQR_outlier_detection(arr)
print(out)
```
