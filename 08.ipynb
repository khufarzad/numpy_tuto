```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('BimMain.csv')
dff = pd.read_csv('BimFil.csv')
print(df.describe())
print(dff.describe())
```

```python
from scipy.stats import skew, kurtosis
```

```python
v = np.array(df['wi'])
```

```python
m = np.mean(v)
s = np.std(v)
sk = skew(v)
ku = kurtosis(v)

print(m)
print(s)
print(sk)
print(ku)
plt.hist(v, bins=200)
plt.show()
```

```python
vf = np.array(dff['wf'])
m = np.mean(vf)
s = np.std(vf)
sk = skew(vf)
ku = kurtosis(vf, fisher=False)

print(m)
print(s)
print(sk)
print(ku)
plt.hist(vf, bins=200)
plt.show()
```

```python
plt.plot(v, lw=0.3)
plt.plot(vf, lw=0.4)
```
