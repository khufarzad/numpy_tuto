```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([7, 13, 5, 7, 1])
```

```python
plt.plot(a, b)
```

```python
x = np.linspace(0, 30, 1000)
y = np.around(np.sin(x), 4)
z = np.around(np.cos(x), 4)
plt.plot(x, y, color='r', label='sin(x)')
plt.plot(x, z, color='k', label='cos(x)')
plt.xlabel('x')
plt.legend(loc='lower right')

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 1, 2, 1])
plt.scatter(x, y)
plt.show()
```

```python
df = pd.read_csv('sig.csv')
u = np.array(df['u'])
v = np.array(df['v'])
w = np.array(df['w'])
```

```python
plt.plot(u, label='u(cm/s)')
plt.plot(v, color='crimson', label='v(cm/s)')
plt.plot(w, color='k', label='w(cm/s)')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.scatter(v, w, c='r', s=2, label='u-v')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(projection='3d')
ax1.scatter(u, v, w, marker='.', label='u-v-w')
plt.show()
```

```python
up = u - np.mean(u) # fluctuation
```

```python
plt.plot(up, color='crimson', label='Signal')
plt.plot([0, len(up)], [np.mean(up), np.mean(up)], color='k', label='Mean')
plt.legend()
plt.show()
```

```python
std_up = np.std(up)
std_u = np.std(u)
print(std_up, std_u)
```
